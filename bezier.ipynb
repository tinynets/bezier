```python
import matplotlib.pyplot as plt
import numpy as np
from Point import Point
```

```python
start_point = Point(15, 10)
control_point = Point(25, 25)
end_point = Point(40, 10)
```

```python
def calc_bezier(t, points):

    p0, p1, p2 = points

    x = (1 - t) ** 2 * p0.x + 2*(1-t)*p1.x*t + t**2*p2.x
    y = (1 - t) ** 2 * p0.y + 2*(1-t)*p1.y*t + t**2*p2.y

    return Point(x, y)



def plot_bezier_curve(start_point, control_point, end_point):
    points = [start_point, control_point, end_point]
    x = []
    y = []
    for t in np.linspace(0, 1, 100):
        p = calc_bezier(t, points)
        x.append(p.x)
        y.append(p.y)

        plt.plot(control_point.x, control_point.y, 'ro')

    plt.plot(x, y)
    plt.show()
    
```

```python
plot_bezier_curve(start_point, control_point, end_point)
```
